```python
from __future__ import absolute_import, division, print_function, unicode_literals

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, Flatten, Dropout, MaxPooling2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import RMSprop
from keras.utils.np_utils import to_categorical 

import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_dir = '../input/arts/train'
validation_dir = '../input/arts/validation'

train_guay_dir = os.path.join(train_dir, 'guay')  # directory with our training guay pictures
train_nielsen_dir = os.path.join(train_dir, 'nielsen')  # directory with our training nielsen pictures
train_avon_dir = os.path.join(train_dir, 'mckinnon')  # directory with our training avon pictures
validation_guay_dir = os.path.join(validation_dir, 'guay')  # directory with our validation guay pictures
validation_nielsen_dir = os.path.join(validation_dir, 'nielsen')  # directory with our validation 
validation_avon_dir = os.path.join(validation_dir, 'mckinnon')  # directory with our validation avon pictures

num_guay_tr = len(os.listdir(train_guay_dir))
num_nielsen_tr = len(os.listdir(train_nielsen_dir))
num_avon_tr = len(os.listdir(train_avon_dir))

num_guay_val = len(os.listdir(validation_guay_dir))
num_nielsen_val = len(os.listdir(validation_nielsen_dir))
num_avon_val = len(os.listdir(validation_avon_dir))

total_train = num_guay_tr + num_nielsen_tr + num_avon_tr 
total_val = num_guay_val + num_nielsen_val + num_avon_val

print('total training guay images:', num_guay_tr)
print('total training nielsen images:', num_nielsen_tr)
print('total training nielsen images:', num_avon_tr)

print('total validation guay images:', num_guay_val)
print('total validation nielsen images:', num_nielsen_val)
print('total validation nielsen images:', num_avon_val)
print("--")
print("Total training images:", total_train)
print("Total validation images:", total_val)
```

```python
batch_size = 80 # 128
epochs = 200 #200
IMG_HEIGHT = 290
IMG_WIDTH = 375
num_classes = 3
```

```python
# This function will plot images in the form of a grid with 1 row and 
# 5 columns where images are placed in each column.
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 5, figsize=(20,20))
    axes = axes.flatten()
    for img, ax in zip( images_arr, axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
image_gen_train = ImageDataGenerator(
                    rescale=1./255,
                    width_shift_range=.15,
                    height_shift_range=.15,
                    horizontal_flip=True,
                    zoom_range=0.5
                    )

train_data_gen = image_gen_train.flow_from_directory(batch_size=batch_size,
                                                     directory=train_dir,
                                                     shuffle=True,
                                                     target_size=(IMG_HEIGHT, IMG_WIDTH),
                                                     class_mode='categorical')
```

```python
train_labels = train_data_gen.classes 

# convert the training labels to categorical vectors 
train_labels = to_categorical(train_labels, num_classes=num_classes)

image_gen_val = ImageDataGenerator(rescale=1./255)

val_data_gen = image_gen_val.flow_from_directory(batch_size=batch_size,
                                                 directory=validation_dir,
                                                 target_size=(IMG_HEIGHT, IMG_WIDTH),
                                                 class_mode='categorical')

val_labels = val_data_gen.classes 
```

```python
# 
# convert the training labels to categorical vectors 
val_labels = to_categorical(val_labels, num_classes=num_classes)

model_new = Sequential([
    Conv2D(8, 3, padding='same', activation='relu', 
           input_shape=(IMG_HEIGHT, IMG_WIDTH ,3)),
    MaxPooling2D(),
    Dropout(0.2),
    Conv2D(16, 3, padding='same', activation='relu'),
    MaxPooling2D(),
    Dropout(0.2),
    Conv2D(32, 3, padding='same', activation='relu'),
    MaxPooling2D(),
    Conv2D(64, 3, padding='same', activation='relu'),
    MaxPooling2D(),
    Dropout(0.2),
    Flatten(),

    Dense(512, activation='relu'),
    Dense(3, activation='softmax')
])

optimizer = RMSprop(0.0005)
# 'RMSprop'
sgd = tf.keras.optimizers.SGD(lr=0.001, decay=1e-6, momentum=0.9, nesterov=True)

# 'adam'

model_new.compile(optimizer=sgd,
                  loss=tf.keras.losses.CategoricalCrossentropy(),
                  metrics=['accuracy'])

model_new.summary()

history = model_new.fit_generator(
    train_data_gen,
    steps_per_epoch=total_train // batch_size,
    epochs=epochs,
    validation_data=val_data_gen,
    validation_steps=total_val // batch_size
)
```

```python
acc = history.history['acc']
val_acc = history.history['val_acc']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(epochs)

plt.figure(figsize=(16, 16))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('200e, 10b, rmsprop(.0005), dense(512)')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

# Evaluate

```python


score = model_new.evaluate_generator(val_data_gen, verbose=1)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
